```python
from pyspark.sql import functions as F

catalog = "kailyn_klaassen"
schema = "odcs"
base_path = "/Volumes/kailyn_klaassen/odcs/energydata"

spark.sql(f"USE CATALOG {catalog}")
spark.sql(f"USE {schema}")

# 1) plant_details (bronze)
plant_df = (
    spark.read
        .format("csv")
        .option("header", "true")
        .option("inferSchema", "true")
        .load(f"{base_path}/plant_details.csv")
)

plant_df.write.mode("overwrite").format("delta").saveAsTable(f"{catalog}.{schema}.plant_details_bronze")

# 2) battery_sensor_feed (bronze)
battery_df = (
    spark.read
        .format("csv")
        .option("header", "true")
        .option("inferSchema", "true")
        .load(f"{base_path}/battery_sensor_feed.csv")
)

battery_df.write.mode("overwrite").format("delta").saveAsTable(f"{catalog}.{schema}.battery_sensor_feed_bronze")

# 3) solar_sensor_feed (bronze)
solar_df = (
    spark.read
        .format("csv")
        .option("header", "true")
        .option("inferSchema", "true")
        .load(f"{base_path}/solar_sensor_feed.csv")
)

solar_df.write.mode("overwrite").format("delta").saveAsTable(f"{catalog}.{schema}.solar_sensor_feed_bronze")

# 4) wind_sensor_feed (bronze)
wind_df = (
    spark.read
        .format("csv")
        .option("header", "true")
        .option("inferSchema", "true")
        .load(f"{base_path}/wind_sensor_feed.csv")
)

wind_df.write.mode("overwrite").format("delta").saveAsTable(f"{catalog}.{schema}.wind_sensor_feed_bronze")

```
