```python
catalog = "kailyn_klaassen"
schema = "odcs"

spark.sql(f"USE CATALOG {catalog}")
spark.sql(f"USE {schema}")

# Plant details
spark.sql(f"""
CREATE OR REPLACE TABLE {catalog}.{schema}.plant_details_silver
AS
SELECT *
FROM {catalog}.{schema}.plant_details_bronze
""")

# Battery sensor feed
spark.sql(f"""
CREATE OR REPLACE TABLE {catalog}.{schema}.battery_sensor_feed_silver
AS
SELECT *
FROM {catalog}.{schema}.battery_sensor_feed_bronze
""")

# Solar sensor feed
spark.sql(f"""
CREATE OR REPLACE TABLE {catalog}.{schema}.solar_sensor_feed_silver
AS
SELECT *
FROM {catalog}.{schema}.solar_sensor_feed_bronze
""")

# Wind sensor feed
spark.sql(f"""
CREATE OR REPLACE TABLE {catalog}.{schema}.wind_sensor_feed_silver
AS
SELECT *
FROM {catalog}.{schema}.wind_sensor_feed_bronze
""")
```
